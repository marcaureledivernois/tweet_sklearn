### Let's loop over 500 steps now!

We denote K the number of steps.
It takes around 6min to compute 500 steps. But you can loop over less steps to reduce computational time.

```python
%%time
# CPU times: total: 9min 54s
# Wall time: 6min 25s

all_weights = pd.DataFrame(columns = ticker)
K = 500

for i in range(K):
    simulated_r = None
    simulated_mu = None
    simulated_cov_mat = None
    
    weights_efficient_frontier = []
    
    for gam in np.linspace(-0.5,1.5,101):
        weights_efficient_frontier.append(efficient_frontier2(None, None, None, None)[2])
        
    to_append = pd.DataFrame(weights_efficient_frontier, columns = ticker)
    to_append['gamma'] = np.linspace(-0.5,1.5,101) # We append gamma as last column of the dataframe. So we now to which gamma the portfolio corresponds to.

    all_weights = pd.concat([all_weights, to_append], ignore_index = True)
```
